```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# BASIC PLOTING 

```python
#line 1 :-
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# resize the graph.

plt.figure(figsize=(5,3) , dpi=150)

# longhand method
plt.plot(x,y , label='2X' , color = "green" , linewidth=2 , marker="." , markeredgecolor="black" , markersize=10 , linestyle="--")
## Shorthand method
#plt.plot(x,y, "g*--" ,label='2X')
plt.title("OUR FIRST GRAPH", fontdict = {'fontname': 'Comic sans MS' , 'fontsize' : 20})
plt.xlabel("X Axis" , fontdict = {'fontname': 'Comic sans MS' , 'fontsize' : 15})
plt.ylabel("Y Axis" , fontdict = {'fontname': 'Comic sans MS' , 'fontsize' : 15})

#line 2 :-
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2,"r",label= "x^2")
plt.plot(x2[5:], x2[5:]**2 , "r--")

# only for projection purpose.
# >>> plt.plot(x2[:4], x2[:4]**2,"r",label= "x^2")


plt.xticks([0,1,2,3,4,])
#plt.yticks([0,5,10,15,20])
plt.legend()
plt.show()
```

# BAR GRAPH

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)

plt.figure(figsize=(5,3))
        

patterns = ["/","o","**"]
for bar in bars :
    bar.set_hatch(patterns.pop())

#bars[0].set_hatch("/")
#bars[1].set_hatch("o")
#bars[2].set_hatch("**")

plt.show()
```
